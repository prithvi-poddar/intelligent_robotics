## A* implementation

In the outputs of the A* algorithm, **grey** nodes will be the nodes that were explored by the algorithm and the **black** line will be the shortest path from the Start (**marked by an x**) to the Goal (**marked by a solid black circle**)

### 5 obstacles environment

```python
from astar import *
```

```python
n_row=10 
n_col=10
n_obs=5
start=[9,0]
goal=[0,9]
n_obs = 5
astar = A_star(n_row=n_row, n_col=n_col, n_obs=n_obs, start=start, goal=goal)
```

```python
astar.generate_path()
```

```python
astar.visualize()
```
